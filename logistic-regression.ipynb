# Regressão logística

__Pacotes utilizados__

* `pandas`: para criação e manipulação de banco de dados
* `numpy`: para trabalhar com algebra linear
* `matplotlib`: para plotar gráficos
* `train_test_split`: para selecionar parte do dataset

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
np.random.seed(42) # para manter os mesmos resultados
```

```
x = np.linspace(15, 42, 15).astype(int)
y_logit = x * -0.5 + 15 + np.random.normal(0, 1.5, x.shape)
y = np.round(np.exp(y_logit) / (np.exp(y_logit) + 1), 0)

plt.scatter(x,y)
plt.tight_layout()
```

```
from sklearn.linear_model import LogisticRegression

x = np.array(x)
y = np.array(y)

log_reg = LogisticRegression()
log_reg.fit(x.reshape(-1,1), y)

y_proba = log_reg.predict_proba(x.reshape(-1,1))

plt.scatter(x,y)
plt.plot(x, y_proba[:, 1], "r-",)
plt.tight_layout()
```
